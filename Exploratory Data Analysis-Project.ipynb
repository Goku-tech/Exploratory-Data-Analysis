<b>Maths Score Preprocessing:

```python
sns.boxplot(x=sp["math score"],color="blue")
```

```python
x=sp["math score"]
Q1=x.quantile(0.25)
print("25 Percentile",Q1)
Q2=x.quantile(0.75)
print("75 Percentile",Q2)
IQR=Q2-Q1
print("Inter Quartile Range",IQR)
print(Q1-1.5*(IQR))
print(Q2+1.5*(IQR))
```

```python
a=sp[sp["math score"]>27]
a["math score"].median()
sp.loc[sp["math score"]<27,"math score"]=np.nan
sp.fillna(a["math score"].median(),inplace=True)
```

<b>After Preprocessing

```python
sns.boxplot(x=sp["math score"],color="blue")
```

<b>Here since there is no value greater than 107 from the graph no need to preprocess (Q2+1.5*(IQR))
